```python
import pandas as pd
import numpy as np
```

```python
mapped_df = pd.read_csv('mapped_data-v2.csv')
mapped_df.head()
```

```python
target = 'DC201'
mapped_df = mapped_df.dropna(subset=target)
```

```python
NUMERICAL_DATA = ['DC216', 'DC220', 'DC142a']
CATEGORICAL_DATA = [column for column in mapped_df.columns if column != target and column not in NUMERICAL_DATA]

numerical_df = mapped_df[NUMERICAL_DATA].copy()
categorical_df = mapped_df[CATEGORICAL_DATA].astype('category').copy()
target_df = mapped_df[target].copy()
```

```python
merged_df = pd.concat([numerical_df, categorical_df, target_df], axis=1)
merged_df
```

```python
merged_df_missing = pd.DataFrame(merged_df.isna().sum().loc[merged_df.isna().sum() > 0], columns=['missing_data_count'])
merged_df_missing['missing_data_type'] = merged_df.dtypes
merged_df_missing['missing_data_percentage'] = merged_df_missing['missing_data_count'] / len(merged_df)
merged_df_missing = merged_df_missing.sort_values(by='missing_data_percentage', ascending=False)
merged_df_missing
```

```python
mapped_df_missing = pd.DataFrame(mapped_df.isna().sum().loc[mapped_df.isna().sum() > 0], columns=['missing_data_count'])
mapped_df_missing['missing_data_type'] = mapped_df.dtypes
mapped_df_missing['missing_data_percentage'] = mapped_df_missing['missing_data_count'] / len(mapped_df)
mapped_df_missing = mapped_df_missing.sort_values(by='missing_data_percentage', ascending=False)
mapped_df_missing
```

```python
import miceforest as mf

# Create kernels.  # Mice Forest
kernel = mf.ImputationKernel(
  data=merged_df,
  save_all_iterations=True,
  random_state=1343
)
# Run the MICE algorithm for 5 iterations on each of the datasets
kernel.mice(5,verbose=True)
#print(kernel)
completed_dataset_mf = kernel.complete_data(dataset=0, inplace=False)
```

```python
pd.DataFrame(completed_dataset_mf).head(15)
```
